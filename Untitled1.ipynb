```python
# Crear una copia del DataFrame
data_numerica = DatoTitanic.copy()

# Convertir columnas de tipo object a numéricas
for column in data_numerica.columns:
    if data_numerica[column].dtype == 'object':
        label_encoder = LabelEncoder()
        data_numerica[column] = label_encoder.fit_transform(data_numerica[column])

# Verificar el resultado
print(data_numerica.head())
```

```python
# Dividir el dataset en Train y Test para evaluar correctamente el modelo.
```

```python
# Dividir los datos en conjunto de entrenamiento y prueba
X = data_numerica.drop('Survived', axis=1)
Y = data_numerica['Survived']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=0)
```

```python
# Imputar NaN en X_train usando la media de las columnas
X_train.fillna(X_train.mean(), inplace=True)
# Imputar NaN en X_test usando la media de las columnas
X_test.fillna(X_train.mean(), inplace=True)
```

```python
#Entrenar el modelo configurando los diferentes hiperparámetros.
```

```python
# Crear y entrenar el modelo de regresión lineal
modelo_regresion = LinearRegression()
modelo_regresion.fit(X_train, Y_train)
```

```python
# Realizar predicciones
Prediccion = modelo_regresion.predict(X_test)
```

```python
#Interpretar, analizar y documentar los resultados obtenidos
```

Coeficientes: Los coeficientes de la regresión indican la relación entre las variables independientes (entradas) y la variable dependiente (salida). Cada valor corresponde a un coeficiente para una característica específica. Los coeficientes más cercanos a cero indican una menor influencia de esas características en el modelo.

Por ejemplo, el coeficiente -1.93e-05 para la primera variable independiente es muy pequeño, lo que sugiere que esa característica tiene poca influencia en la predicción. En cambio, el coeficiente -0.5116 para otra variable tiene una influencia considerable en la variable dependiente.
Error cuadrático medio (ECM): El error cuadrático medio es 0.1391, lo que indica la magnitud promedio de los errores entre las predicciones del modelo y los valores reales. Un valor más bajo de ECM indica un mejor ajuste del modelo, pero no se puede juzgar de manera aislada sin un contexto o comparación con otros modelos. Dependiendo de la escala de los datos, este valor podría ser razonable o necesitar mejorarse.

Coeficiente de determinación (R²): El R² de 0.4075 indica que aproximadamente el 40.75% de la variabilidad en la variable dependiente es explicada por el modelo. Este valor sugiere que el modelo tiene un ajuste moderado, pero hay una parte significativa de la variabilidad que no está siendo capturada por el modelo. Un valor de R² cercano a 1.0 indica un buen ajuste, mientras que un valor cercano a 0.0 indica un ajuste pobre.
